```python
import pandas as pd
import numpy as np
from wordcloud import WordCloud

file = 'FB_Data.xlsx'
data = pd.ExcelFile(file)
df_sheet1 = data.parse('Amarjargal')
df = df_sheet1.iloc[:,[10]]
df_copy = df.copy()

```

```python
df_copy.head()
```

```python
text =' '.join(df_copy['message'].astype(str))
```

```python
import matplotlib.pyplot as plt
```

```python
my_cloud = WordCloud(background_color='White',stopwords=['nan','Амжилт','хүсье','amjilt','Amjilt']).generate(text)
plt.imshow(my_cloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
