```python
import pandas as pd
import numpy as np
from wordcloud import WordCloud
import matplotlib.pyplot as plt

file = 'election_sample_data.xlsx'
data = pd.ExcelFile(file)
df_sheet1 = data.parse(0)

```

```python
df_sheet1.head()
```

```python
df_sheet1['RT']=df_sheet1['text'].str.contains('RT')
```

```python
df_sheet1.head()
```

```python
df= df_sheet1[df_sheet1['RT']==False]
```

```python
df.head()
```

```python
text =' '.join(df['text'].astype(str))
```

```python
# break the string into list of words 
str_list = text.split() 

# gives set of unique words 
unique_words = set(str_list) 

for words in unique_words : 
    print('Frequency of ', words , 'is :', str_list.count(words)) 
```

```python

my_cloud = WordCloud(stopwords=['–Ω—ç—Ä','https'], background_color='White').generate(text)
plt.imshow(my_cloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
